```python
import pandas as pd
import gmaps
from config import api_key
# Import File (CSV)
latlon_df=pd.read_csv("../Resources/states_lat_lon.csv")

latlon_df.head()
```

```python
latlon_df = latlon_df.drop([0,1])
```

```python
latlon_df = latlon_df.drop(41)
```

```python
#latlon_df=latlon_df.set_index(['Unnamed: 3'])
len(latlon_df)
display(latlon_df)
```

```python
latlon_df=latlon_df.rename(columns={'Unnamed: 1': 'Latitude','Unnamed: 2': 'Longitude','Unnamed: 3':'state'})
display(latlon_df)
```

```python
latlon_df=latlon_df[['Latitude','Longitude','state']]
latlon_df.head()
```

```python
locations = latlon_df[["Latitude", "Longitude"]]
len(locations)
```

```python
breweries_df=pd.read_csv('../Resources/breweries.csv')
breweries_df
```

```python
breweries_df.value_counts()
```

```python
brewery_count=breweries_df.value_counts()
brewery_count.dtypes
```

```python
merged_set=pd.merge(latlon_df,breweries_df,on='state')
merged_set.dropna()
```

```python
state_count=merged_set['state'].value_counts()
state_count
```

```python
len(locations)
df = pd.DataFrame(state_count)
state = df["state"]
state
```

```python
locations = locations.astype(float)
state = state.astype(float)
```

```python
# Plot Heatmap
fig = gmaps.figure()


# Create heat layer
heat_layer = gmaps.heatmap_layer(locations[["Latitude", "Longitude"]], weights=state, 
                                 dissipating=False, max_intensity=100,
                                 point_radius=2)


# Add layer
fig.add_layer(heat_layer)

# # Display figure
fig
```
